```python
from tensorflow import keras
```

```python
from keras.datasets import imdb
```

```python
(X_train,y_train),(X_test,y_test)=imdb.load_data() 
```

```python
from keras.preprocessing.text import Tokenizer
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
padded_train=pad_sequences(X_train)
padded_test=pad_sequences(X_test,maxlen=2494)
```

```python
print(padded_train.shape)
print(padded_test.shape)
```

```python
model=keras.models.Sequential([
    keras.layers.Embedding(200000, 64, input_length=2494),
    keras.layers.Flatten(),
    keras.layers.Dense(1,activation='sigmoid'),
    
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(padded_train,y_train,epochs=1)
```

```python
acc=model.evaluate(padded_test,y_test) 
```

```python
print('loss: ',acc[0])
print('accuracy: ',acc[1])
```
